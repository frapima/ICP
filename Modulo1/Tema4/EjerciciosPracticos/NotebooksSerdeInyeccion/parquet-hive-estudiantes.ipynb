```python
%%writefile parquethive.hql

create  table parquet_datacleaninternal (
  idref  STRING,   
  ident  STRING,
  gsm19023 FLOAT,  
  gsd19024 FLOAT, 
  gsd19025 FLOAT,    
  gsd19026 FLOAT, 
  genetitle STRING,   
  genesymbol STRING,
  geneID SMALLINT,    
  uniGenetitle STRING,
  uniGenesymbol STRING,   
  uniGeneID STRING,
  NucleotideTitle STRING) 
STORED AS PARQUET;
```
